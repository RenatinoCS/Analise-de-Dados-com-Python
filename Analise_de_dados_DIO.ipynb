# **2 - Documentos das capitais.xlsx**

```python
dfaracaju = pd.read_excel('Aracaju.xlsx')
dffortaleza = pd.read_excel('Fortaleza.xlsx')
dfnatal = pd.read_excel('Natal.xlsx')
dfrecife = pd.read_excel('Recife.xlsx')
dfsalvador = pd.read_excel('Salvador.xlsx')
```

```python
#juntando os arquivos
dfcapitais = pd.concat([dfaracaju,dffortaleza,dfnatal,dfrecife,dfsalvador])
dfcapitais.head()
```

```python
#pegar uma amostra de dados aleatoria
dfcapitais.sample(10)
```

```python
dfcapitais.dtypes #TIPOS DE DADOS
```

```python
#Mudar o tipo de dado LojaID prar object, pois não vai ser utilizado para calculos
dfcapitais['LojaID'] = dfcapitais['LojaID'].astype('object')
dfcapitais.dtypes
```

```python
#verificar valores faltantes
dfcapitais.isnull().sum()
```

**Criando colunas novas**

```python
#Criando a coluna de receita
dfcapitais['Receita'] = dfcapitais['Vendas'].mul(dfcapitais['Qtde'])
dfcapitais.head()
```

# **3 - Trabalahndo com datas**

```python
#nem sempre a coluna de data está no tipo data, então tem que verificar e mudar o tipo caso precise

```

```python
#Agrupando a receita por ano
dfcapitais.groupby(dfcapitais['Data'].dt.year)['Receita'].sum() #dt = data tipe
```

```python
#Criar coluna ano
dfcapitais['Ano_Venda'] = dfcapitais['Data'].dt.year
dfcapitais.sample(8)
```

```python
#acressentar coluna mês e dia em uma coluna
dfcapitais['Mes_Venda'], dfcapitais['Dia_Venda'] = (dfcapitais['Data'].dt.month, dfcapitais['Data'].dt.day)
dfcapitais.sample(8)
```

```python
#data mais antiga
dfcapitais['Data'].min()
```

```python
#diferença entre datas, utilizando a data mais aintiga como inicial
dfcapitais['Diferenca_Dias'] = dfcapitais['Data'] - dfcapitais['Data'].min()
dfcapitais.sample(8)
```

```python
#criar coluna trimestre
dfcapitais['Trimestre_Venda'] = dfcapitais['Data'].dt.quarter
dfcapitais.sample(8)
```

```python
#filtrar vendas de 2019 do mês março
vendas_marco_19 = dfcapitais.loc[(dfcapitais['Data'].dt.year == 2019) & (dfcapitais['Data'].dt.month == 3)]
vendas_marco_19.sample(8)
```

```python
#excluir coluna data
dfcapitais.drop(['Data'], axis=1, inplace=True)
dfcapitais.head()
```

#  5 - Análise exploratoria

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
#Criando o dataFrame

df2 = pd.read_excel('AdventureWorks.xlsx')
```

```python
df2.head(3)
```

```python
#Quantidade de linhas e colunas
df2.shape
```

```python
df2.dtypes
```

```python
#Qual a receita total
df2['Valor Venda'].sum()
```

```python
#Qual o custo total ? Primeiro cira a coluna custo
df2['custo'] = df2['Custo Unitário'].mul(df2['Quantidade']) #Criando a coluna de custo
df2.head(3)
```

```python
#Agora calcula o total do custo
round(df2['custo'].sum(),2)
```

```python
#Lucro = Valor da venda - custo
df2['lucro'] = df2['Valor Venda'] - df2['custo']
df2.head(3)
```

```python
#lucro total
round(df2['lucro'].sum(),2)
```

```python
#Coluna para saber o tempo de envio de de pois da compra
df2['Tempo_envio'] = df2['Data Envio'] - df2['Data Venda']
df2.head(3)
```

# **Descobrindo o tempo de envio para cada marca, para isso é precisso transformar a coluna Tempo_envio em númerica negrito**

```python
#Extraindo apenas os dias
df2['Tempo_envio'] = (df2['Data Envio'] - df2['Data Venda']).dt.days
df2.head(3)
```

```python
#verificando se ficou int
df2['Tempo_envio'].dtypes
```

```python
#Media do tempo de envio por Marca
df2.groupby('Marca')['Tempo_envio'].mean() #groupby agrupa, []filtra +/-
```

```python
#Verificando se tem valores ausentes nas colunas
df2.isnull().sum()
```

```python
#Agrupando por ano e marca
pd.options.display.float_format = '{:20,.2f}'.format#Para os numeros não ficarem em conotação cientifica
df2.groupby([df2['Data Venda'].dt.year, 'Marca'])['lucro'].sum()
```

```python
#armazenando a infomação a cima em uma variavel resetando o index
lucro_ano = df2.groupby([df2['Data Venda'].dt.year, 'Marca'])['lucro'].sum().reset_index()
lucro_ano
```

```python
#Total de produtos vendidos
df2.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False)#nofinal é para ficar em ordem
```

```python
#Grafico Total de produtos vendidos
df2.groupby('Produto')['Quantidade'].sum().sort_values(ascending=True).plot.barh(title='Total de produtos vendidos')
plt.xlabel('Total')
plt.ylabel('Produto');
```

```python
#Grafico Lucro por ano
df2.groupby(df2['Data Venda'].dt.year)['lucro'].sum().plot.bar(title='Lucro por ano')
plt.xlabel('Ano')
plt.ylabel('Receita');
```

```python
#Codigo para ver o resultado indicado no grafico Lucro por ano
df2.groupby(df2['Data Venda'].dt.year)['lucro'].sum()
```

```python
#Selecionando apenas as vendas do ano de 2009
df2_2009 = df2[df2['Data Venda'].dt.year == 2009]
df2_2009.head(2)
```

```python
#Grafico de linhas lugro por mês
df2_2009.groupby(df2_2009['Data Venda'].dt.month)['lucro'].sum().plot(title='Lucro x Mês')
plt.xlabel('Mês')
plt.ylabel('Lucro');
```

```python
#Grafico de linhas lugro por marca
df2_2009.groupby('Marca')['lucro'].sum().plot.bar( color='green', title= 'Lucro x Marca')
plt.xlabel('Marca')
plt.ylabel('Lucro');
plt.xticks(rotation='horizontal') #Rotaciona a legenta x
```

```python
#Grafico Lucro por classe
df2_2009.groupby('Classe')['lucro'].sum().plot.bar( color='green', title= 'Lucro x Classe')
plt.xlabel('Classe')
plt.ylabel('Lucro');
plt.xticks(rotation='horizontal')
```
